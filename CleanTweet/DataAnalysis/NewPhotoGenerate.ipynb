```python
def my_autopct(pct):
    print(pct)
    return (int(pct)+'%') if (pct-int(pct) == 0) else ('%.2f%%' % pct)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,10))


vaderPosCount = 716843 
vaderNegCount = 576462 
vaderNeuCount = 177904
covidInPos = 9854
covidInNeg =  9426
covidInNeu = 1891

labels = ['Positive', 'Negative', 'Neutral']

size = 0.47
vals = np.array([
    [vaderPosCount - covidInPos, covidInPos], 
    [vaderNegCount - covidInNeg, covidInNeg], 
    [vaderNeuCount - covidInNeu, covidInNeu]
])

cmap = plt.colormaps["tab20c"]
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap([1, 2, 5, 6, 9, 10])

# ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors,
#        wedgeprops=dict(width=size, edgecolor='w'))

# ax.pie(vals.flatten(), radius=1-size, colors=inner_colors,
#        wedgeprops=dict(width=size, edgecolor='w'))

# ax.set(aspect="equal")




_, _, autotexts = ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'), labels = labels, autopct= my_autopct, textprops={'fontsize': 14})
for autotext in autotexts:
    autotext.set_color('white')
plt.setp(autotexts, size=14, weight="bold")
    
_, _, autotexts = ax.pie(vals.flatten(), radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='w'), labels = ['', 'COVID-19', '','COVID-19', '','COVID-19'], autopct=my_autopct, textprops={'fontsize': 14})

ax.set(aspect="equal")
#fig = plt.figure(figsize =(10, 7))
#_, _, autotexts = plt.pie(data, labels = cars, autopct='%1.1f%%')
for autotext in autotexts:
    autotext.set_color('teal')

plt.setp(autotexts, size=12, weight="bold")


plt.savefig("sentiWithCovidTemp.png",dpi=1080, bbox_inches='tight')
plt.show()
```
